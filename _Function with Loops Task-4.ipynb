Problem 1: Mobile Recharge Validation System

 Problem Statement:
Create a function to validate a mobile recharge request.

Rules:
Recharge amount must be ≥ ₹50


Valid plans: 199, 299, 399, 599


User should be allowed to retry if invalid


Real-Time Use
Telecom recharge apps


Payment gateways


Hint
Use function for validation


Use while loop for retry




```python
def recharge(amount, plan):
  if(amount>=50):
    if(plan==199 or plan==299 or plan==399 or plan==59):
      print("Recharge Successful")
    else:
      print("Invalid Plan")
  else:
    print("Invalid Amount")
while True:
  amount=int(input("Enter the amount"))
  plan=int(input("Enter the plan"))
  recharge(amount,plan)
  if(input("Do you want to continue ")=="no"):
    break

```

Problem 2: Inventory Reorder Alert System

Problem Statement:
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15


Print Stock OK otherwise


Real-Time Use
Warehouse systems


Retail management


Hint
Use function


Use for loop on dictionary


```python
def check_inventory(products):
    reorder = 15
    print("--- Inventory Status ---")
    for product, stock in products.items():
        if stock < reorder:
            print(f"{product}: Reorder Alert (Current Stock: {stock})")
        else:
            print(f"{product}: Stock OK (Current Stock: {stock})")
product_inventory = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 5,
    "Monitor": 18,
    "Webcam": 12,
    "Printer": 30
}
check_inventory(product_inventory)
```

Problem 3: Student Result Processing System

Problem Statement:
Create a function that:
Accepts a list of student marks


Calculates average marks


Prints Pass if average ≥ 50 else Fail


Real-Time Use
Online exam systems


Evaluation portals


Hint
Use loop to calculate total


Return result from function



```python
def process_student_result(marks):
    if not marks:
        print("No marks provided. Cannot calculate average.")
        return

    total_marks = sum(marks)
    average_marks = total_marks / len(marks)

    print(f"Student Marks: {marks}")
    print(f"Average Marks: {average_marks:.2f}")

    if average_marks >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")
print("--- Student Result Processing ---")

student1_marks = [60, 75, 80, 55, 70]
print("\nProcessing Student 1:")
process_student_result(student1_marks)

student2_marks = [40, 30, 45, 50, 25]
print("\nProcessing Student 2:")
process_student_result(student2_marks)

student3_marks = [50, 50, 50, 50, 50]
print("\nProcessing Student 3:")
process_student_result(student3_marks)

student4_marks = []
print("\nProcessing Student 4:")
process_student_result(student4_marks)
```

Problem 4: Cab Fare Estimator with Retry Option

Problem Statement:
Create a cab fare calculator:
Base fare = ₹50


₹12 per km


Peak hour → 25% extra


Allow retry using while loop


Real-Time Use
Ride-hailing apps


Transport systems


Hint
Function for fare calculation


While loop for retry





```python
def calculate_cab_fare(distance_km, is_peak_hour):
    base_fare = 50
    fare_per_km = 12
    peak_hour_surcharge_rate = 0.25

    initial_fare = base_fare + (distance_km * fare_per_km)

    if is_peak_hour:
        total_fare = initial_fare * (1 + peak_hour_surcharge_rate)
    else:
        total_fare = initial_fare

    return total_fare

print("Welcome to the Cab Fare Estimator!")

while True:
    try:
        distance_str = input("Enter distance traveled in kilometers: ")
        distance_km = float(distance_str)

        if distance_km < 0:
            print("Distance cannot be negative. Please try again.\n")
            continue

        peak_hour_input = input("Is it a peak hour? (yes/no): ").lower()
        is_peak_hour = False
        if peak_hour_input == 'yes':
            is_peak_hour = True
        elif peak_hour_input == 'no':
            is_peak_hour = False
        else:
            print("Invalid input for peak hour. Please enter 'yes' or 'no'. Try again.\n")
            continue

        fare = calculate_cab_fare(distance_km, is_peak_hour)
        print(f"Estimated Cab Fare: ₹{fare:.2f}")
        break
    except ValueError:
        print("Invalid input. Please enter a number for distance. Try again.\n")
    except Exception as e:
        print(f"An unexpected error occurred: {e}. Please try again.\n")
```

Problem 5: Employee Attendance Eligibility Checker

Problem Statement:
Create a function that:
Accepts attendance list ("P", "A")


Calculates attendance percentage


Returns Eligible if ≥ 75% else Not Eligible


Real-Time Use
HR payroll systems


Internship tracking


 Hint
Use loop to count present days


Use return statement




```python
def check_attendance_eligibility(attendance_list):
    if not attendance_list:
        return "No attendance recorded."

    total_days = len(attendance_list)
    present_days = attendance_list.count('P')

    attendance_percentage = (present_days / total_days) * 100

    print(f"Attendance: {attendance_list}")
    print(f"Present Days: {present_days} out of {total_days}")
    print(f"Attendance Percentage: {attendance_percentage:.2f}%")

    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

print("--- Employee Attendance Eligibility Checker ---")

attendance1 = ['P', 'P', 'A', 'P', 'P', 'P', 'A', 'P', 'P', 'P'] # 80% attendance
print("\nEmployee 1:")
print(f"Status: {check_attendance_eligibility(attendance1)}")

attendance2 = ['P', 'A', 'A', 'P', 'P', 'A', 'P', 'A', 'P', 'A'] # 50% attendance
print("\nEmployee 2:")
print(f"Status: {check_attendance_eligibility(attendance2)}")

attendance3 = ['P', 'P', 'P', 'P', 'P', 'P', 'P', 'A'] # 7 out of 8 = 87.5%, 6 out of 8 = 75%
print("\nEmployee 3:")
print(f"Status: {check_attendance_eligibility(['P', 'P', 'P', 'P', 'P', 'P', 'A', 'A'])}")

attendance4 = []
print("\nEmployee 4:")
print(f"Status: {check_attendance_eligibility(attendance4)}")
```

Problem 6: Password Strength Checker

Problem Statement:
Create a function to check password strength.
Rules:
Minimum length = 8


Must contain at least one digit


Must contain at least one special character (@#$)


Real-Time Use
Login systems


Security validation


Hint
Use loop over string


Use conditions inside function




```python
def check_password_strength(password):
    min_length = 8
    has_digit = False
    has_special_char = False
    special_chars = "@#$"

    if len(password) < min_length:
        return "Weak: Password must be at least 8 characters long."

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special_char = True

    if not has_digit:
        return "Weak: Password must contain at least one digit."
    if not has_special_char:
        return "Weak: Password must contain at least one special character (@#$)."

    return "Strong: Password meets all criteria."

print("--- Password Strength Checker ---")

password_test1 = "MySecureP@ss1"
print(f"Password '{password_test1}': {check_password_strength(password_test1)}")

password_test2 = "short@1"
print(f"Password '{password_test2}': {check_password_strength(password_test2)}")

password_test3 = "NoSpecialChar123"
print(f"Password '{password_test3}': {check_password_strength(password_test3)}")

password_test4 = "NoDigit!@#$"
print(f"Password '{password_test4}': {check_password_strength(password_test4)}")

password_test5 = "verylongpasswordwithallcondition123!@#"
print(f"Password '{password_test5}': {check_password_strength(password_test5)}")
```
