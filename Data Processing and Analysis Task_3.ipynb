### Problem Statement 1: Employee Performance Bonus Eligibility


```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

def bonus_eligible_employees(employees):
    highest_score = max(employees.values())
    top_performers = [name for name, score in employees.items() if score == highest_score]
    print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")

bonus_eligible_employees(employees)

```

### Problem Statement 2: Search Query Keyword Analysis

```python
query = "Buy mobile phone buy phone online"

def keyword_analysis(sentence):
    sentence = sentence.lower()
    words = sentence.split()

    result = {}

    for word in words:
        if words.count(word) > 1:
            result[word] = words.count(word)

    return result


keyword_analysis(query)
```

### Problem Statement 3: Sensor Data Validation



```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

def sensor_data_validation(lst):
    final_output = []

    for hour, value in enumerate(lst):
        if value % 2 == 0:
            final_output.append((hour, value))

    print(f"Valid Sensor Readings (Hour, Value):\n{final_output}")

sensor_data_validation(sensor_readings)
```

### Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

def email_domain_analysis(emails):
    domains = []

    # Extract domains
    for email in emails:
        parts = email.split("@")
        domains.append(parts[1])

    total = len(domains)

    unique_domains = []

    # Find unique domains
    for domain in domains:
        if domain not in unique_domains:
            unique_domains.append(domain)

    for domain in unique_domains:
        count = domains.count(domain)
        percentage = (count / total) * 100
        print(domain + ":", str(int(percentage)) + "%")

email_domain_analysis(emails)
```

### Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.3
for day, sale in enumerate(sales, start=1):
    if sale >= threshold:  
        print(f"Day {day}: {sale}")
```

### Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

def duplicate_userId(user_ids):
    duplicates = []
    for user in user_ids:
        if user not in duplicates and user_ids.count(user) > 1:
            duplicates.append(user)

    for user in duplicates:
        print(user, "→", user_ids.count(user), "times")

duplicate_userId(user_ids)
```
